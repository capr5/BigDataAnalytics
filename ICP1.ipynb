```python
path='/content/gdrive/MyDrive/Data.csv'
df_data= pd.read_csv(path)
df_data.head(10)
```

```python
df_data.columns
```

```python
#checking for null values 
df_data.isnull().sum()
```

```python
data_mal = df_data[df_data.diagnosis == "M"]
data_ben = df_data[df_data.diagnosis == "B"]
```

```python
#plotting a pie chart to identify the distribution of data based on diagnosis(Benign or Malignamt cases)
labels=['Malignant','Benign']
values= [len(data_mal),len(data_ben)]
trace_data = [go.Pie(labels=labels, values=values, marker=dict(colors=['red','green']))]
layout_data = go.Layout(title="Percentage of Malignant(M) and Benign(B) cases")
figure_data= go.Figure(data=trace_data,layout= layout_data)
pyo.iplot(figure_data)
```

```python
#The below plot shows the correlation between mean values of texture,radius, concavity, perimiter,area and smoothness for specific diagnosis type(B or M)
sns.pairplot(df_data,hue = 'diagnosis', palette= 'pastel', vars = ['radius_mean', 'texture_mean','concavity_mean', 'perimeter_mean','area_mean','smoothness_mean'])
```

```python
#detecting the outliers using box plot for the feature "radius_mean" for both type of diagnosis which avoid resulting in a less accurate model after training
sns.boxplot(x=df_data['radius_mean'], y=df_data['diagnosis'])
```

```python
##detecting the outliers using box plot for the feature "texture_mean" for both type of diagnosis which avoid resulting in a less accurate model after training
sns.boxplot(x=df_data['texture_mean'], y=df_data['diagnosis'])
```
